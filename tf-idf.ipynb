```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
import pandas as pd
import numpy as np
from google.colab import drive
```

```python
drive.mount('/content/drive')
df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Dataset_final.xlsx')
```

```python
X = df.loc[:, 'review'].to_numpy()
y = df.loc[:, 'attribute'].to_numpy()

X_train = ['']*5928
y_train = ['']*5928

for i in range(5928):
  X_train[i] += str(X[i])
  y_train[i] += str(y[i])

print(X_train)
```

```python
vectorizer = TfidfVectorizer(input=X_train)
X_train = vectorizer.fit_transform(X_train)
```

```python
df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Trainingskorpus_Final.xlsx')

X = df.loc[:, 'data'].to_numpy()
y = df.loc[:, 'final_Label'].to_numpy()

X_test = ['']*1438
y_test = ['']*1438

for i in range(1438):
  X_test[i] += str(X[i])
  y_test[i] += str(y[i])

X_test = vectorizer.transform(X_test)
```

```python

lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)


right = 0
for i in range(1438):
  if(y_pred[i] == y_test[i]):
    right += 1


print("Accuracy on test set (\"Trainingskorpus\") is ", right*100./1438, "%")
```

```python

y_pred = lr.predict(X_train)


right = 0
for i in range(5928):
  if(y_pred[i] == y_train[i]):
    right += 1


print("Accuracy on training set (\"Dataset\") is ", right*100./5928, "%")
```
