```python
# merge both datasets on data column
df = pd.merge(df_trainingskorpus_M, df_trainingskorpus_A, on='data')
df
```

```python
# drop irrelevant columns
df = df[['data','label 1','Unnamed: 15']]
df
```

```python
# rename column
df = df.rename(columns={'Unnamed: 15': 'label 2'})

# remove null values
df = df.dropna(how='any',axis=0) 

# remove all duplicate reviews
df=df.drop_duplicates(subset=['data'])
df

# drop rows with two labels
a = ['B' , 'P', 'D', 'I']
df = df[df['label 2'].isin(a)]
df
```

```python
print(df['label 1'].value_counts())
```

```python
print(df['label 2'].value_counts())
```

```python
# calculate cohens kappa between rater 1 and rater 2
from sklearn.metrics import cohen_kappa_score

cohen_kappa_score(df['label 1'], df['label 2'])
```
